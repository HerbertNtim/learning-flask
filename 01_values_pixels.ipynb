```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array in the range(0, 255), 16 x 16
arr = np.arange(0, 256, 1)
arr
```

```python
arr = arr.reshape(16, 16)
arr
```

```python
arr_8bits = arr.astype(np.uint8)
```

```python
plt.imshow(arr_8bits, cmap='gray')
plt.axis('off')
plt.show();
```
